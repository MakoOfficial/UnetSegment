```python
import os
import cv2
import numpy as np
from PIL import Image

def resize_origin(mask_dir, data_dir, target_dir):
    os.makedirs(target_dir,exist_ok=True)
    masks = os.listdir(mask_dir)
    for i in masks:
        print(f"precessing pic {i}")
        img_path = os.path.join(data_dir, i)
        img = Image.open(img_path)
        h, w = img.size
        img = img.resize((int(h/1.5), int(w/1.5)))
        save_path = os.path.join(target_dir, i)
        img.save(save_path)  
```

```python
mask_dir = "../archiveMasked/mask_train/test/"
data_dir = "../archive/boneage-training-dataset/"
target_dir = "../archiveMasked/mask_train/test_origin"

resize_origin(mask_dir, data_dir, target_dir)
```

```python
###########  Resize ######################
from PIL import ImageOps
class resize:
    def __init__(self, reshape_size=512):
        self.reshape_size = reshape_size
        pass

    def __call__(self, img):
        w, h = img.size
        long = max(w, h)
        # 按比例缩放成512
        w, h = int(w / long * self.reshape_size), int(h / long * self.reshape_size)
        # 压缩并插值
        img = img.resize((w, h))
        # 然后是给短边扩充，使用ImageOps.expand
        delta_w, delta_h = self.reshape_size - w, self.reshape_size - h
        padding = (delta_w // 2, delta_h // 2, delta_w - (delta_w // 2), delta_h - (delta_h // 2))
        img = ImageOps.expand(img, padding)
        return img

def pad_resize(data_dir , mask_dir, target_dir):
    os.makedirs(target_dir,exist_ok=True)
    masks = os.listdir(mask_dir)
    mask_save = os.path.join(target_dir, "mask")
    os.makedirs(mask_save,exist_ok=True)
    input_save = os.path.join(target_dir, "input")
    os.makedirs(input_save, exist_ok=True)
    for i in masks:
        print(f"precessing pic {i}")
        img_path = os.path.join(data_dir, i)
        mask_path = os.path.join(mask_dir, i)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
        img = Image.fromarray(np.uint8(img))
        mask = Image.fromarray(np.uint8(mask))
        
        resizer = resize(512)
        img_resize = resizer(img)
        mask_resize = resizer(mask)
        
        img_resize.save(os.path.join(input_save, i))  
        mask_resize.save(os.path.join(mask_save, i))
```

```python
data_dir = '../archiveMasked/UnetDataset/train_unpadding/input/'
mask_dir = '../archiveMasked/UnetDataset/train_unpadding/mask/'
target_dir = '../archiveMasked/UnetDataset/train/'

pad_resize(data_dir, mask_dir, target_dir)
```
